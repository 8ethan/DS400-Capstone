```python
import time
import os
import pandas as pd
import numpy as np
from pcap_parse import *
```

```python
normal_dir = "D:\\Notebooks\\Datasets\\Network_dataset_pcaps\\normal_pcaps"
mal_dir = "D:\\Notebooks\\Datasets\\Network_dataset_pcaps\\normal_attack_pcaps"
truth_dir = "D:\\Notebooks\\Datasets\\truth"
```

```python
files = os.listdir(truth_dir)
truth_df = pd.DataFrame()
for dataset in files:
    cur = pd.read_csv(truth_dir+"\\"+dataset)
    truth_df = pd.concat([truth_df, cur], axis=0)
truth_df.sort_values(by=['ts','src_ip','dst_ip','src_port','dst_port'], inplace=True)
truth_df.reset_index(drop=True, inplace=True)
```

```python
df = pd.DataFrame(columns=["p_bytes","label","ts","src_ip","dst_ip","src_port","dst_port","pcap_ver"])
#df = parse_pcap(normal_dir+"\\normal_IoT_2.pcap", df, 0)
#df = parse_pcap(mal_dir+"\\normal_scanning\\normal_scanning1.pcap", df, 0)
for file_name in os.listdir(mal_dir+"\\normal_scanning"):    
    df = parse_pcap(mal_dir+"\\normal_scanning\\"+file_name, df, 1)
df.sort_values(by=['ts','src_ip','dst_ip','src_port','dst_port'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
scan_truth = truth_df[truth_df['type']=='scanning']
scan_truth.sort_values(by=['ts','src_ip','dst_ip','src_port','dst_port'], inplace=True)
scan_truth.reset_index(drop=True, inplace=True)
```

```python
cur_idx = 0
base_ts = 0
no_match = []
is_truth = np.zeros(len(df))
for index, row in scan_truth.iterrows():
    
    base_ts = row['ts']
    if index%10000==0:
        print(index)
    backup_idx = cur_idx
    
    while cur_idx < len(df):
        
        if row['ts'] == df.loc[cur_idx]['ts']:
            if row['src_ip'] == df.loc[cur_idx]['src_ip'] and row['dst_ip'] == df.loc[cur_idx]['dst_ip']:
                if row['src_port'] == df.loc[cur_idx]['src_port'] and row['dst_port'] == df.loc[cur_idx]['dst_port']:
                    #print("Row "+str(index)+" matches Row "+str(cur_idx))
                    is_truth[cur_idx] = 1
                    cur_idx+=1
                    break
        if base_ts < df.loc[cur_idx]['ts']:
            #print("No match for row "+str(index))
            no_match.append(index)
            cur_idx = backup_idx
            break
        cur_idx+=1
```

```python
df['val_label'] = is_truth
```
